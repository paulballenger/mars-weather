```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime

%matplotlib inline
```

```python
bna = pd.read_csv('../data/BNA_weather.csv')
bna.head()
```

```python
#only keep needed columns
bna = bna[['ls', 'TMIN','TMAX']].copy()
bna.head()
```

```python
bna.info()
```

```python
#group by ls number and get mean temps

bna_1 = bna.groupby(['ls']).mean().round(2)
bna_1 = bna_1.reset_index()

#rename columns
bna_1 = bna_1.rename(columns = {'TMIN':'bna_min', 'TMAX':'bna_max'})

bna_1.tail()
```

```python
bna_1['bna_avg'] = bna_1[['bna_min', 'bna_max']].mean(axis=1).round(2)
bna_1.head()
```

```python
mars = pd.read_csv('../data/mars_weather.csv')
mars.head()
```

```python
#only keep needed columns
mars = mars[['ls', 'min_temp', 'max_temp']].copy()
mars.head()
```

```python
mars.info()
```

```python
#group by ls number and get mean temps

mars_1 = mars.groupby(['ls']).mean().round(2)
mars_1 = mars_1.reset_index()

#rename columns
mars_1 = mars_1.rename(columns = {'min_temp':'mars_min', 'max_temp':'mars_max'})

mars_1.tail()
```

```python
mars_1['mars_avg'] = mars_1[['mars_min', 'mars_max']].mean(axis=1).round(2)
mars_1.head()
```

```python
#merge df's
temps = pd.merge(bna_1, mars_1, on = 'ls', how = 'inner')
temps.head()
```

```python
temps.to_csv('temps.csv', index = False)
```
